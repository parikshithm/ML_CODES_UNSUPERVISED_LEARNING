```python
#Selected columns
selected_columns_df = raw_df
```

```python
##Import the required libraries
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
##%pip install mlxtend
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
#-------------
##First set the minimum support
##This gives the rules that meet the min.support criteria
##--------------------
minimum_required_support=0.1
```

```python
##Apply association mining algorithm using this support
frequent_itemsets = apriori(selected_columns_df, min_support=minimum_required_support, use_colnames=True)
```

```python
frequent_itemsets
```

```python
## get rules from the frequent itemsets that macth the desired lift
## set the required lift
minimum_required_lift = 1.1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=minimum_required_lift)
```

```python
print("Total number of rules generated by the algorithm with given support and lift:", len(rules))
```

```python
#Set the minimum threshold for confidence
minimum_required_confidence = 0.5
```

```python
#Filter rules baased on confidence and lift thresholds
filtered_rules = rules[(rules['confidence'] >= minimum_required_confidence)]
```

```python
filtered_rules
```

```python
print("Total number of filtered rule that meet the lift, confidence and support criteria:", len(filtered_rules))
```

```python
#Sorting the rules based on lift, confidence and support criteria (all in descending order)
sorted_rules = filtered_rules.sort_values(by=['lift', 'confidence', 'support'], ascending=False)
```

```python
sorted_rules
```

```python
##Selecting the top 5 values
top_5_rules = sorted_rules.head(5)
```
